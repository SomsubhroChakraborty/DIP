enhance image using image negative


```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/cartoon.jpeg")

# Convert into grayscale image
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Get height and width
row, column = img.shape[:2]  # Corrected variable name from "image" to "img"

# Create a new image array
new_image = np.zeros([row, column], dtype=np.uint8)  # Corrected dtype from np.unit8 to np.uint8

# Reverse the intensity using a for loop
for i in range(row):
    for j in range(column):
        new_image[i, j] = 255 - img[i, j][0]  # Reversed intensity for each channel

# Display images
cv2_imshow(new_image)
cv2_imshow(img)
cv2_imshow(grey)
```

Histogram equilization process


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def his_equalization(image):
    H, W = image.shape

    # Calculate the original histogram
    histogram = np.zeros(256, dtype=int)
    for i in range(H):
        for j in range(W):
            histogram[image[i, j]] += 1

    # Plot the original histogram
    plt.subplot(2, 1, 1)
    plt.bar(np.arange(len(histogram)), histogram)
    plt.title('Original Histogram')

    # Calculate CDF
    cdf = np.zeros(256, dtype=float)
    cdf[0] = histogram[0] / (H * W)
    for i in range(1, 256):
        cdf[i] = cdf[i - 1] + histogram[i] / (H * W)

    equalized = np.round(cdf[image] * 255).astype(np.uint8)

    # Calculate the equalized histogram
    equalized_histogram = np.zeros(256, dtype=int)
    for i in range(H):
        for j in range(W):
            equalized_histogram[equalized[i, j]] += 1

    # Plot the equalized histogram
    plt.subplot(2, 1, 2)
    plt.bar(np.arange(len(equalized_histogram)), equalized_histogram)
    plt.title('Equalized Histogram')

    plt.show()

    return equalized

# Read the input image
img = cv2.imread('/content/cartoon.jpeg', cv2.IMREAD_GRAYSCALE)
img = img.astype(np.uint8)

# Perform histogram equalization
equalized_image = his_equalization(img)

# Display the original and equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()


```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/cartoon.jpeg',0)

gls=np.zeros_like(img)
glsbg=np.zeros_like(img)
(r,c)=img.shape[:2]
a=120
b=150
for i in range(r):
    for j in range(c):
        if (img[i,j]>=a and img[i,j]<=b):
            gls[i,j]=255
            glsbg[i,j]=255
        else:
            gls[i,j]=0
            glsbg[i,j]=img[i,j]
gls=gls.astype(np.uint8)
glsbg=np.array(glsbg,'uint8')

cv2_imshow(gls)
cv2_imshow(glsbg)

```

```python
#Salt and pepper noise and median filtering
import cv2
import numpy as np

from google.colab.patches import cv2_imshow

img = cv2.imread('/content/cartoon.jpeg',0)

snp=np.random.normal(0,9,img.size)
snp=snp*0.5
snp=snp.reshape(img.shape[0],img.shape[1]).astype(np.uint8)

noiseImg=img+(img*snp)
cv2_imshow(img)
cv2_imshow(noiseImg)

def median_filter(img):
    r,c=img.shape
    dst=np.zeros((r,c),dtype=np.uint8)
    for i in range(1,r-1):
        for j in range(1,c-1):
            temp=[img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
            temp.sort()
            dst[i,j]=temp[9//2]
    return dst
# dst=cv2.medianBlur(noiseImg,5)
nimg=median_filter(img)
cv2_imshow(nimg)
# cv2_imshow(dst)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/cartoon.jpeg', 0)

def mean_filter(img):
    dst = np.zeros_like(img)  # Initialize dst inside the function
    kernel = np.ones([3, 3], dtype=np.float32) / 9
    (r, c) = img.shape[:2]
    for i in range(1, r - 1):
        for j in range(1, c - 1):
            dst[i, j] = np.sum(img[i - 1:i + 2, j - 1:j + 2] * kernel)
            dst[i, j] = np.clip(dst[i, j], 0, 255)
    dst = dst.astype(np.uint8)
    return dst

new_img = mean_filter(img)

cv2_imshow(img)
cv2_imshow(new_img)




```
